<h3 class="fa fa-bookmark-o"> 项目名称：对外数据接口自动化</h3>

+ 负责无线全景算法表中 LTE FDD 制式指标算法的需求文档、 LLD 设计文档的评审和算法文档测试用例设计及编写，准备测试数据包，根据实际情况编写 SQL 脚本进行算法验证，使用TMSS 管理并追踪缺陷，对测试过程进行回顾整理；后续引入自动化流程进行算法和接口的自动化测试；
+ 利用 iBatis 框架完成了LTE FDD 下各个指标的SQL 脚本执行后的数据验证，后期将此方法推广至UMTS、LTE TDD、GSM 制式所有指标的 SQL 脚本执行与数据正确新性验证，提升近50%的测试效率；
+ 使用 Postman 对数据接口业务测试，执行接口用例，发现 BUG 并使用TMSS 进行缺陷的追踪，对测试过程进行回顾整理。使用 SoupUI 对接口的并发性能和响应进行性能方面的测试，并输出测试报告提供异常指标给开发提供优化；
+ 后期通过引入 iBatis + TestNg 框架，完成了对外 17 个数据接口自动化测试搭建，并创造多种场景的测试数据(接口参数的边界值、数据类型的容错性、缺省参数、默认值、大数据量返回值)，对接口进行自动化测试。从而保证对外接口的准确性和稳健性。

通过自动化测试流程的引入，指标算法的测试减少了人工重复执行 SQL 脚本并核对数据的过程，整体效率提升了近 50%；接口自动化的引入，减少了人工手动赋值进行接口调用和数据核对重复劳动，节省了测试周期去关注接口容错性、性能方面的测试。同时自身提升了 SQL 测试脚本编写、自动化接口测试和接口的性能测试的能力。
